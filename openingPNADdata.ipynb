## Opening PNAD Data

```python
#The example below selects four variables just as an example. Usually to analyze the data someone would probably select 20-50 
#variables, but it's much better renaming 20-50 variables than about 400 variables. And there's the need to rename 
#them only once. 
```

```python
pnad4thtrimester2020selectedcolumns = pnad4thtrimester2020[['V2001','V2003','V2005','V2007']]
```

```python
pnad4thtrimester2020selectedcolumns.rename({'V2001':'numberofhouseholds', 'V2003': 'ordernumber','V2005': 'relationtohousehold', 'V2007':'sex'}, axis = 1, inplace = True)
```
