[Rais 2019](http://www.rais.gov.br/sitio/index.jsf) stands for Relação Anual de Informações Sociais (Annual Social Information List). The microdata below is from rais2019, and one of the data's column is [CBO](https://concla.ibge.gov.br/classificacoes/por-tema/ocupacao/classificacao-brasileira-de-ocupacoes), which stands for Classificação Brasileira de Ocupações (Classification of Occupations). The Brazilian Classification of Occupations describes and orders occupations within a hierarchical structure that allows for the aggregation of information regarding the workforce. The data comes with CBO's Occupation that is more specific, I wanted to turn it into a CBO Broad Group and Main SubGroup with its respective names. That is what this code does. The data dictionary comes from [here](http://www.mtecbo.gov.br/cbosite/pages/downloads.jsf#)
